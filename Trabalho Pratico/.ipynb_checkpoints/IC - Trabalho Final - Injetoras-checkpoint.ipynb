```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(rc = { "figure.figsize": (10, 7) })
sns.set_style("white")
%matplotlib inline
```

```python
# Import data
df = pd.read_csv("tcs-data.csv")

# Convert milliseconds to datetime
# df = df.drop(["time_str"], axis = 1)
# df["datetime"] = pd.to_datetime(df["time"], unit = "ms")

df
```

```python
# Remove redundant values, like two consecutive 1s or 0s
redundant_indexes = []
previous_state = df.injecting[0]

for i, state in enumerate(df.injecting[1:]):
    if state == previous_state:
        # Index i is shifted in -1
        redundant_indexes.append(i + 1)

    previous_state = state

df = df.drop(redundant_indexes).reset_index(drop = True)
df
```

```python
# Get the deltas between each row
state_changes = df.time.diff().shift(-1)
state_changes
```

```python
# Get the deltas from injecting = 1 to next injecting = 1
cycles = df[(df.injecting == 1)].time.diff().shift(-1)
cycles
```

```python
# Removing ends and aligning data
cycles = cycles[:-1]
state_changes = state_changes[1:-2:2]
```

```python
# Calculate cooling time
dataset = pd.concat([cycles, state_changes], axis = 1, keys = ["cycle", "injecting"])
dataset["cooling"] = dataset.cycle - dataset.injecting
dataset
```

```python
# Remove cycles longer than 2 minutes
dataset = dataset[dataset.cycle < 120000]
dataset
```

```python
# Show data distribution
plt.hist(dataset, bins = 50, stacked = True)
plt.legend(["cycle", "injecting", "cooling"])
plt.xlabel("Time (ms)")
plt.ylabel("Occurrences")
plt.title("Data Distribution")
plt.show()
```

```python
# Plot Injecting vs Total Cycle
plt.scatter(dataset.cycle, dataset.injecting, s = 2)
plt.xlabel("Total Cycle (ms)")
plt.ylabel("Injecting (ms)")
plt.title("Injecting vs Total Cycle")
plt.show()
```

```python
# Standardize data
std_cycle = (dataset.cycle - dataset.cycle.mean()) / dataset.cycle.std()
std_injecting = (dataset.injecting - dataset.injecting.mean()) / dataset.injecting.std()
```

```python
# Plot Standardized Injecting vs Standardized Total Cycle
plt.scatter(std_cycle, std_injecting, s = 2)
plt.xlabel("Standardized Total Cycle")
plt.ylabel("Standardized Injecting")
plt.title("Standardized Injecting vs Standardized Total Cycle")
plt.show()
```

```python
# Since there were no significant changes after standardization, use original data to maintain units

# Calculate inertias for 100 KMeans models
inertias = []
for i in range(1, 101):
    kmeans = KMeans(n_clusters = i, init = "k-means++", n_init = "auto").fit(dataset[["cycle", "injecting"]])
    inertias.append(kmeans.inertia_)
```

```python
# Plot Inertia vs Quantity of Clusters
plt.plot(range(1, 101), inertias)
plt.xticks(np.arange(1, 101, 5))
plt.xlabel("Quantity of Clusters")
plt.ylabel("Inertia")
plt.title("Inertia vs Quantity of Clusters")
plt.show()
```

```python
# Zoom in on Inertia vs Quantity of Clusters
plt.plot(range(10, 101), inertias[9:])
plt.xticks(np.arange(10, 101, 5))
plt.xlabel("Quantity of Clusters")
plt.ylabel("Inertia")
plt.title("Inertia vs Quantity of Clusters (Zoom)")
plt.show()
```

```python
# After analysing, select 55 clusters
kmeans_55 = KMeans(n_clusters = 55, init = "k-means++", n_init = "auto").fit(dataset[["cycle", "injecting"]])
dataset.insert(3, "KMeans-55", kmeans_55.predict(dataset[["cycle", "injecting"]]))
dataset
```

```python
# Get the center of each cluster
clusters = pd.DataFrame(kmeans_55.cluster_centers_, columns = ["cycle", "injecting"])
clusters["cooling"] = clusters.cycle - clusters.injecting
clusters.head()
```

```python
# Plot Injecting vs Total Cycle (With Clusters)
sns.scatterplot(
    dataset[["cycle", "injecting", "KMeans-55"]], x = "cycle", y = "injecting", hue = "KMeans-55",
    palette = sns.color_palette(n_colors = 55), legend = False, s = 6
).set(
    title = "Injecting vs Total Cycle (With Clusters)",
    xlabel = "Total Cycle (ms)",
    ylabel = "Injecting (ms)"
)

sns.scatterplot(clusters, x = "cycle", y = "injecting", marker="X", color = "black")
plt.show()
```

```python
# Plot Injecting vs Total Cycle (With Clusters and Zoom)
zoomed_data = dataset[["cycle", "injecting", "KMeans-55"]][dataset.cycle < 42000]
sns.scatterplot(
    zoomed_data, x = "cycle", y = "injecting", hue = "KMeans-55",
    palette = sns.color_palette(n_colors = len(pd.unique(zoomed_data["KMeans-55"]))), legend = False, s = 20
).set(
    title = "Injecting vs Total Cycle (With Clusters and Zoom)",
    xlabel = "Total Cycle (ms)",
    ylabel = "Injecting (ms)"
)

sns.scatterplot(clusters[clusters.cycle < 42000], x = "cycle", y = "injecting", marker="X", s = 75, color = "black")
plt.show()
```

```python
# Plot Injecting vs Cooling (With Clusters)
sns.scatterplot(
    dataset[["cooling", "injecting", "KMeans-55"]], x = "cooling", y = "injecting", hue = "KMeans-55",
    palette = sns.color_palette(n_colors = 55), legend = False, s = 6
).set(
    title = "Injecting vs Cooling (With Clusters)",
    xlabel = "Cooling (ms)",
    ylabel = "Injecting (ms)"
)

sns.scatterplot(clusters, x = "cooling", y = "injecting", marker="X", color = "black")
plt.show()
```

```python
# Plot Injecting vs Cooling (With Clusters and Zoom)
zoomed_data = dataset[["cooling", "injecting", "KMeans-55"]][dataset.cycle < 42000]
sns.scatterplot(
    zoomed_data, x = "cooling", y = "injecting", hue = "KMeans-55",
    palette = sns.color_palette(n_colors = len(pd.unique(zoomed_data["KMeans-55"]))), legend = False, s = 20
).set(
    title = "Injecting vs Cooling (With Clusters and Zoom)",
    xlabel = "Cooling (ms)",
    ylabel = "Injecting (ms)"
)

sns.scatterplot(clusters[clusters.cycle < 42000], x = "cooling", y = "injecting", marker="X", s = 75, color = "black")
plt.show()
```
