```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# dados de apartamentos e seus preços
df2 = pd.read_csv("data2.txt", names=['tamanho', 'quartos', 'preco'])
```

```python
%matplotlib notebook
# Creating figure
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")

# Creating plot
ax.scatter3D(df2.tamanho, df2.quartos, df2.preco, color = "green")
plt.title("Tamanho, Quartos e Preços de Casas")
ax.set_xlabel('Tamanho m2')
ax.set_ylabel('Quartos')
ax.set_zlabel('Preço')

# show plot
plt.show()
```

## Standardization vs Normalization

```python
# standardization scales the mean to zero, no particular range
df2 = (df2 - df2.mean())/df2.std()

# normalization scales the values from 0 to 1
# df2 = (df2 - df2.min())/(df2.max()-df2.min())
```

```python
t0, t1, t2 = np.random.random(3)
lr = 0.1
epoch = 1000
m = df2.size
for _ in range(epoch):
    j = 0
    derivada_j_t0 = 0
    derivada_j_t1 = 0
    derivada_j_t2 = 0
    for i, x in enumerate(zip(df2.tamanho, df2.quartos)):
        h = t0 + t1*x[0] + t2*x[1]
        j += (h - df2.preco[i])**2
        derivada_j_t0 += h - df2.preco[i]
        derivada_j_t1 += (h - df2.preco[i])*x[0]
        derivada_j_t2 += (h - df2.preco[i])*x[1]
    j /= (2*m)
    derivada_j_t0 /= m
    derivada_j_t1 /= m
    derivada_j_t2 /= m

    t0 -= lr*derivada_j_t0
    t1 -= lr*derivada_j_t1
    t2 -= lr*derivada_j_t2

    print(j)
```

```python
# Create the figure
fig = plt.figure()

# Add axes
ax = fig.add_subplot(111,projection='3d')

# plot the regression plane
X, Y = np.meshgrid(df2.tamanho, df2.quartos)
Z = t0 + t1*X + t2*Y
ax.plot_surface(X, Y, Z, alpha=0.05, color='red')
ax.scatter3D(df2.tamanho, df2.quartos, df2.preco)
```
