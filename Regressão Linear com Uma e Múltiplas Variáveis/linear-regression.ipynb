```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# foodtrucks em diferentes cidades
df1 = pd.read_csv("data1.txt", names=['populacao', 'lucro'])
# dados de apartamentos e seus pre√ßos
df2 = pd.read_csv("data2.txt", names=['tamanho', 'quartos', 'preco'])
```

```python
df1.plot.scatter('populacao', 'lucro', marker='x')
```

```python
t0, t1 = np.random.random(2)
lr = 0.01
epoch = 2000
m = df1.size
for _ in range(epoch):
    j = 0
    derivada_j_t0 = 0
    derivada_j_t1 = 0
    for i, x in enumerate(df1.populacao):
        h = t0 + t1*x
        j += (h - df1.lucro[i])**2
        derivada_j_t0 += h - df1.lucro[i]
        derivada_j_t1 += (h - df1.lucro[i])*x
    j /= (2*m)
    derivada_j_t0 /= m
    derivada_j_t1 /= m

    t0 -= lr*derivada_j_t0
    t1 -= lr*derivada_j_t1
```

```python
df1.plot.scatter('populacao', 'lucro', marker='x')
plt.plot([5, 22.5], [t0+t1*5, t0+t1*22.5], c='red')
```

```python
%matplotlib notebook
# Creating figure
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")

# Creating plot
ax.scatter3D(df2.tamanho, df2.quartos, df2.preco, color = "green")
plt.title("Tamanho Quartos e Pre√ßos de Apartamentos")
ax.set_xlabel('Tamanho m2')
ax.set_ylabel('Quartos')
ax.set_zlabel('Pre√ßo')

# show plot
plt.show()
```
